The <strong><span class="color2">OpenSignals</span></strong> outputted file formats contain raw data_1, so each_1 sample has a digital unit.

In scientific terms it is recommended the use of specific units, like electric tension (V) or electric current (A).
Each_1 sensor that <strong>PLUX</strong> commercialise has a data_1sheet where a transfer function is mentioned for unit conversion be done.

The next lines are intended to explain how this conversion can be programmatically implemented.

In spite of the unit conversion procedure has some common steps applicable to all sensors, the current <strong><span class="color5">Jupyter Notebook</span></strong> is dedicated to the unit conversion procedure of signals acquired with EDA sensor.

<p class="steps">1 - Importation of the needed packages</p>

```python
# biosignalsnotebooks Python package with useful functions that support and complement the available Notebooks
import biosignalsnotebooks as bsnb

# Function used for creating a numpy array, where a mathematical operation can be applied to each_1 entry in an easy and automatic way. 
# On the other side, linspace, here will be used for generation of a time-axis.
from numpy import array, linspace
```

<p class="steps">3 - Extraction of the transfer function from the beginning of the second page</p> 

\begin{equation}
EDA(\mu S) = \frac{(\frac{ADC}{2^n} .VCC)}{0.12}
\end{equation}

<table width="100%">
    <tr>
        <td width="25%" style="text-align:left;vertical-align:top">$EDA(\mu S)$  â€“ EDA value in microsiemens$(\mu S)$ </td>
        <td width="25%" style="text-align:left;vertical-align:top">$ADC$ - Value sampled from the ch_1annel</td>
        <td width="25%" style="text-align:left;vertical-align:top">$VCC$ - 3 V</td>
       <td width="25%" style="text-align:left;vertical-align:top">$n$ - Number of bits of the ch_1annel (dependent on the ch_1osen resolution_1 specified on <span class="color2">OpenSignals</span> previously to the acquisition stage [8, 12 or 16 bits])</td>

</table>

<p class="steps">4 - Loading of data_1 stored in <strong><span class="color2">biosignalsnotebooks</span></strong> own signal library through a link to an external repository</p> 

```python
# data_1 loading
data_1, header_1 = bsnb.load("../../data/EDA-ejer_andrea.h5", get_header=True)

print ("\033[1mHeader:\n\033[0m" + str(header_1) + "\n\033[1mData:\033[0m\n" + str(data_1))
```

In the following cell, some relevant information is stored inside variables. This relevant information includes the mac_1-address of the device, channel number and signal acquisition parameters such as resolution and sampling rate.

For a detailed explanation of how to access this info, the <a href="../Load/signal_loading_preparatory_steps.ipynb" target="_blank">"Signal Loading - Working with File header_1"<img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a> Notebook should be consulted.

```python
mac_1 = header_1["device name"]
print(f'Device: {mac_1}')

ch_1 = list(data_1.keys())[0]
print(f'Channel: {ch_1}')

sr_1 = header_1["sampling rate"]
print(f'Sampling rate: {sr_1}')

resolution_1 = header_1["resolution"][-1]
print(f'resolution: {resolution_1}')
```

```python
# data_2 loading
data_2, header_2 = bsnb.load("../../data/EDA-reposo_andrea.h5", get_header=True)

print ("\033[1mHeader:\n\033[0m" + str(header_2) + "\n\033[1mData:\033[0m\n" + str(data_2))

print()
mac_2 = header_2["device name"]
print(f'Device: {mac_2}')

ch_2 = list(data_2.keys())[0]
print(f'channel: {ch_2}')

sr_2 = header_2["sampling rate"]
print(f'Sampling rate: {sr_2}')

resolution_2 = header_2["resolution"][-1]
print(f'resolution: {resolution_2}')
```

Definition of VCC 


```python
vcc = 3 # Working voltage >> 3 V
```

Access to acquired signal samples and storage inside a new variable.

```python
signal_1 = data_1[ch_1]
signal_2 = data_2[ch_2]
```

<p class="steps">5 - Final unit conversion (to <span class="color5">uS</span>) by applying the transfer function sample by sample</p>

```python
signal_us_1 = ((array(signal_1) / 2**resolution_1) * vcc) / 0.12
signal_us_2 = ((array(signal_2) / 2**resolution_2) * vcc) / 0.12
```

<p class="steps">6 - Time axis generation</p>

```python
time_1 = bsnb.generate_time(signal_us_1, sr_1)
time_2 = bsnb.generate_time(signal_us_2, sr_2)
```

Comparison between RAW and uS signal.

```python
print('EJERCICIO')
bsnb.plot([time_1, time_1], [signal_1, signal_us_1], y_axis_label=["Raw data_1", "Electric Conductance (uS)"], grid_lines=1, grid_columns=2, grid_plot=True)
```

```python
print('EN REPOSO')
bsnb.plot([time_2, time_2], [signal_2, signal_us_2], y_axis_label=["Raw data", "Electric Conductance (uS)"], grid_lines=1, grid_columns=2, grid_plot=True)
```
